```python
def cek_angka(a, b, c):
    if a != b and a != c and b != c:
        if (a + b == c) or (a + c == b) or (b + c == a):
            return True
    return False

print(cek_angka(2, 3, 5))
print(cek_angka(4, 4, 8))
print(cek_angka(1, 2, 4))
print(cek_angka(7, 2, 5))
```

```python
def cek_digit_belakang(a, b, c):
    a_satuan = a % 10
    b_satuan = b % 10
    c_satuan = c % 10

    if (a_satuan == b_satuan) or (a_satuan == c_satuan) or (b_satuan == c_satuan):
        return True
    return False

a = int(input("Masukkan bilangan pertama: "))
b = int(input("Masukkan bilangan kedua: "))
c = int(input("Masukkan bilangan ketiga: "))

print("Hasil:", cek_digit_belakang(a, b, c))

print(cek_digit_belakang(30, 20, 18))
print(cek_digit_belakang(145, 5, 100))
print(cek_digit_belakang(71, 187, 18))
print(cek_digit_belakang(1024, 14, 94))
print(cek_digit_belakang(53, 8900, 658))
```
